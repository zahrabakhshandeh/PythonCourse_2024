```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Plot

```python
x = np.array([10, 5, 17])
y = np.array([3, 16, 20])
plt.figure(figsize=(10, 8))
plt.plot(x, y, label="XY", c="red", linewidth=3, linestyle=":", 
         marker="o", ms=16, markeredgecolor='blue')
plt.title("My First Plot", c="g", size=24)
plt.xlabel("X", c="r", size=20)
plt.ylabel("Y", c="r", size=20)
plt.legend()
plt.savefig("plot1.png")
plt.show()
```

```python
x = np.linspace(-20, 20, num=20)
# x, x*5, x**2 
plt.figure(figsize=(10, 8))
plt.plot(x, x, label="x")
plt.plot(x, x*5, label="5x")
plt.plot(x, x**2, label="x^2")
plt.legend()
plt.show()
x
```

### Scatter

```python
x = np.array([1, 6, 102, 505, 170, 20, 200, 10])
y = np.array([10, 80, 2, 105, 6, 5, 200, 40])
size = np.array([1, 60, 102, 505, 170, 200, 200, 10])
plt.figure(figsize=(10, 8))
plt.scatter(x, y, c="r", s=y)
plt.show()
```

```python
x1 = np.array([1, 3, 5, 6, 7])
y1 = np.array([0, 12, 4, 5, 10])
x2 = np.array([10, 3, 13, 6, 17])
y2 = np.array([0, 2, 4, 9, 10])
plt.figure(figsize=(10, 8))
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
```

```python
x1 = np.array([1, 3, 5, 6, 7])
y1 = np.array([0, 12, 4, 5, 10])
c = np.array(["blue", "red", "black", "green", "pink"])
plt.figure(figsize=(10, 8))
plt.scatter(x1, y1, c=c)
plt.show()
```

### Pie Plot

```python
x = np.array([20, 30, 120])
y = np.array(["A", "B", "C"])
explode = [0, 0.1, 0]
plt.figure(figsize=(12, 10))
plt.pie(x, labels=y, autopct="%1.2f%%", explode=explode)
plt.legend()
plt.show()
```

### Bar

```python
x = np.array(["PYTHON", "Java", "C++"])
y = np.array([20, 40, 120])
plt.figure(figsize=(10, 8))
plt.bar(x, y, width=0.3)
plt.show()
```

```python
x = np.array(["A", "B", "C"])
y = np.array([20, 40, 120]) 
plt.figure(figsize=(10, 8))
plt.barh(x, y, height=0.3)
plt.show()
```

### Histogram

```python
age = np.array([20, 12, 23, 20, 18, 10, 2, 60])
plt.figure(figsize=(10, 8))
plt.hist(age)
plt.show()
```

```python
data = np.random.normal(1, 100, size=1200)
plt.figure(figsize=(10, 8))
plt.hist(data)
plt.show()
```

### Box Plot

```python
x = np.array([10, 20, 40, 12, 30, -300, 300, 30, 10, 40, 50, -30, -20, -10])
```

```python
plt.figure(figsize=(10, 8))
plt.boxplot(x)
plt.show()
```

### Subplot

```python
x = np.array([0, 2, 3, 4])
y = np.array([4, 3, 4, 10])

plt.figure(figsize=(12, 10))
plt.subplot(2, 3, 1)
plt.plot(x, y)

plt.subplot(2, 3, 2)
plt.pie(x)

plt.subplot(2, 3, 3)
plt.plot(y, x)

plt.subplot(2, 3, 4)
plt.scatter(x, y)

plt.subplot(2, 3, 5)
plt.pie(y)

plt.subplot(2, 3, 6)
plt.boxplot(x)
plt.show()
```

## 3D

```python
ax = plt.axes(projection="3d")
x = np.arange(0, 30, 0.2)
y = np.arange(30, 60, 0.2)
z = np.sin(x) * np.cos(y)
ax.plot(x, y, z)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
